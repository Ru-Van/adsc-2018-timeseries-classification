```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

# Load the Data

```python
with open("data.pickle", "rb") as f:
    data = pickle.load(f)
```

```python
X = data["X"]
y = data["y"]
names = data["names"]
```

# Visualize the data

```python
def show(i):
    print(f"Observation #{i} (Id {names[i]}):")
    pd.DataFrame(X[i].T).plot(subplots=True, figsize=(10, 60))
    plt.show()
```

```python
def show_single(i, j):
    print(f"Observation #{i} (Id {names[i]}) - series #{j}:")
    plt.figure(figsize=(12, 2))
    plt.plot(X[i, j])
    plt.show()
```

## Time series of healthy factories

```python
for i in np.random.choice(np.where(y == 0)[0], size=3, replace=False):
    show(i)
```

## Time series of heavy smoker factories

### Peaks

```python
for i, j in [(59, 4), (59, 31), (249, 34), (237, 20), (245, 7)]:
    show_single(i, j)
```

### Noises

```python
for i, j in [(61, 45), (61, 4), (188, 46), (113, 45), (129, 2)]:
    show_single(i, j)
```

### Constants

```python
for i, j in [(168, 52), (101, 38), (168, 52), (79, 16)]:
    show_single(i, j)
```

### "Special cases"

```python
for i, j in [(201, 40), (201, 54), (103, 29)]:
    show_single(i, j)
```
